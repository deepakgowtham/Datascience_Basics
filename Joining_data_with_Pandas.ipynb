# Inner Join
- only returns rows with matching values on both the dataframes
- df1.merge(df2, on='col1')
- if the values are same but column names are different then use left_on='col1', right_on= 'column'
- if both dataframe have same column name the output contains suffix _x, _y 
- to override use suffixes keyword with tuple as input

image.png

```
taxi_own_veh= taxi_own.merge(taxi_veh, on='vid', suffixes=('_own','_veh'))
taxi_own_veh.head()
```

```
filter_criteria =((ridership_cal_station['month']==7) & 
                 (ridership_cal_station['day_type']=='Weekday') &
                 (ridership_cal_station['station_name']=='Wilson'))
```

```
# here we are using a boolean filter on loc method to filter columns
ridership_cal_station.loc[filter_criteria, 'rides'].sum()
```

# Filtering Joins
- All the above joins discussed are mutating joins,
- Mutatingjoins: Combines data from two tables based on matching observations in both tables
- Filtering joins: Filter observations from table based on whether or not they match an observation in another table
- Filtering joins are Semi join, Anti join

## Semi join
- similar to inner join but returns data only from left table and not right table
- No duplicate rows are returned
- **isin** command

image.png

## Anti Join
- Returns the data from left table without the data from right table
- only data from left table is returned
- new_df= df1.merge(df2, **indicator= True**)
- new_df.loc[ new_df[_merge]== 'left_only', 'col1]

image.png

# Concat
- join tables vertically
- pd.concat([df1, df2, df3])
- ignore_index= True --- resets the index if false same index from each df is retained
- keys =[] -- creates a multi index with names for each table
- sort= True -- sorts column names 
- if one df contains additional columns it is included by defalut to avoid that we can use **join='inner'** this brings only matching columns

# append
- simplified version of concat
- ignore_index, sort == Supported.
- keys, join not supported, join is always outer

## Merge_asof
- similar to ordered merge 
- but instead of work only on exact match it can match with approximate values(less than or equal to) from right table
- in below example 5 is matched with 3 as it is the closest value

- the match criteria can be changed from less than equal to to greater than equal to by **direction='forward'** keyword
default value of direction is backward

- direction='nearest' -- get the nearest values
