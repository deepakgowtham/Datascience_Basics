# Relplot
- relational plot -- can bring up **scatterplot or line plot**

- scatter= when we think each value is independent observation
- line == when we think the same thing varies over time
- **row or col** == seperate plots by row and column by the paramter

```
sns.relplot(x='total_bill',y='tip',data=tips, kind='scatter')
plt.show()
```

```
sns.relplot(x='total_bill',y='tip',data=tips, kind='scatter', col='smoker')
plt.show()
```

## Scatter

### Row and col combined

```
sns.relplot(x='total_bill',y='tip',data=tips, kind='scatter', col='smoker', row='time')
plt.show()
```

```
# col_wrap == limits the number of columns
#col_order,row_order == list of order in which we want to print
sns.relplot(x='total_bill',y='tip',data=tips, kind='scatter', col='day', col_wrap=2,
            col_order=['Thur', 'Fri', 'Sat','Sun'])
plt.show()
```

## Box plot

## Point plot
- show mean of the continous variable for each category
- kind= 'point'
- join = False to remove the interconnecting lines
- estimator= np.median ===to use median instead of mean 
- capsize=0.2 == to add caps at the end of the confidence interval lines

**line plot vs point plot**
- line plot both varaibles are continous
- point both x axis is categorical

**bar plot vs point plot**
- when comparing against other categories using hue parameter it is easier with point plot
- slope of the change between categories is cleary visible with point plot

# Facetting
- Seaborn's FacetGrid is the foundation for building data-aware grids. A data-aware grid allows you to create a series of small plots that can be useful for understanding complex data relationships.
- Multiple plots with shared x and y axis
- also called trellis or lattice plots
- Idea is to quickly to get the trend or idea from different plots when they are side by side
- Data must be tidy to make these plots




## FacetGrid
- Seaborn function for craeting facetgrid
- 2 steps process - defining a facet and mapping the plot type

- g = sns.FacetGrid(df, col="HIGHDEG")
- g.map(sns.boxplot, 'Tuition',       order=['1', '2', '3', '4'])
- first line create a facegrid for the dataframe seperated by columns
- second line map it to box plot for the tuition data in particular order


```
g2= sns.FacetGrid(college, row="Degree_Type",
             row_order=['Graduate', 'Bachelors', 'Associates', 'Certificate'])
g2.map(sns.pointplot, 'SAT_AVG_ALL')
g2.fig.set_size_inches(15,6)
plt.show()
```

## Factorplot
-Seaborn's factorplot() can be a simpler way to create a FacetGrid. Instead of creating a grid and mapping the plot, we can use the factorplot() to create a plot with one line of code
- can automate facetting using Factorplot
- sns.factorplot(x="Tuition", data=df,               col="HIGHDEG", kind='box')


```
g=sns.factorplot(data=college, x='SAT_AVG_ALL', kind='point', row='Degree_Type',
        row_order=['Graduate', 'Bachelors', 'Associates', 'Certificate'])
g.fig.set_size_inches(16,6)
plt.show()
```

```
sns.lmplot(data=college,
        x='SAT_AVG_ALL',
        y='Tuition',
        col="Ownership",
        row='Degree_Type',
        row_order=['Graduate', 'Bachelors'],
        hue='WOMENONLY',
        col_order=['Public', 'Private non-profit'])

plt.show()
plt.clf()
```

# Pairs

## Pairgrid
- Similar to FacetGrid used for visualising the relationship between two variables
- g2 = g.map_diag(plt.hist)
- g3 = g2.map_offdiag(plt.scatter) -- to change the main diagonal and diagonal plots



```
g=sns.PairGrid(data=ins_prem, vars=["fatal_collisions", "premiums"])
g2 = g.map(plt.scatter)

plt.show()
```

```
g = sns.PairGrid(data=ins_prem, vars=["fatal_collisions", "premiums"])
g2 = g.map_diag(plt.hist)
g3 = g2.map_offdiag(plt.scatter)

plt.show()
plt.clf()
```

## Pairplot
- automated way for Pairgrid

```
sns.pairplot(data=ins_prem,
        vars=["fatal_collisions", "premiums"],
        kind='scatter',
        hue='Region',
        palette='RdBu',
        diag_kws={'alpha':.5})
plt.show()
```

```
# adding additonal variables
sns.pairplot(data=ins_prem,
        x_vars=["fatal_collisions_speeding", "fatal_collisions_alc"],
        y_vars=['premiums', 'insurance_losses'],
        kind='scatter',
        hue='Region',
        palette='husl')

plt.show()
plt.clf()
```

```
#using kind to control the types of plots.
# We can also use diag_kind to control the types of plots shown on the diagonals
sns.pairplot(data=ins_prem,
             vars=["insurance_losses", "premiums"],
             kind='reg',
             palette='BrBG',
             diag_kind = 'kde',
             hue='Region')
plt.show()
```

# Joint Grid and Joint Plot
- Used to see relationship between two variables
- JointGrid combines univariate plots such as histograms, rug plots and kde plots with bivariate plots such as scatter and regression plots. 
- Combining multiple plots together

## Joint Grid
- Simple way but has more customization options
- g = sns.JointGrid(data=df, x="Tuition", y="ADM_RATE_ALL") 
- g.plot(sns.regplot, sns.distplot)


## Joint point
- Automated way with less customisations

```
g = sns.JointGrid(x="hum",
            y="total_rentals",
            data=bike,
            xlim=(0.1, 1.0)) 

g.plot(sns.regplot, sns.distplot)
```

```
median=rent['fmr_2'].median() 
mean=rent['fmr_2'].mean() 
median
```

```
fig, ax=plt.subplots()
sns.distplot(rent['fmr_2'], ax=ax)
ax.set(xlabel="1 Bedroom Fair Market Rent",
       xlim=(100,1500),
       title="US Rent")
ax.axvline(x=median, color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=mean, color='b', label='Mean', linestyle='-', linewidth=2)
ax.legend()
plt.show()
```

```
fig, (ax0, ax1)= plt.subplots(nrows=1,ncols=2, sharey=True)
sns.distplot(rent['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))
sns.distplot(rent['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))
plt.show()
```
